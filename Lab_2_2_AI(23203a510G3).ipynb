### Task 1: Cleaning Sensor Data - Filtering Negative Values

To clean the IoT sensor data, we will create a Python function to filter out all negative numbers from a given list, as they are considered invalid. We will then demonstrate its usage with a sample list.

````python
# Sample sensor data list with positive and negative values
sensor_data = [10, -5, 20, 0, -15, 30, -2, 25]

print("Original sensor data:", sensor_data)
````

````python
def filter_negative_numbers(data_list):
    """
    Filters out all negative numbers from a list.

    Args:
        data_list (list): A list of numerical values.

    Returns:
        list: A new list containing only non-negative numbers.
    """
    return [num for num in data_list if num >= 0]

# Apply the function to the sensor data
cleaned_data = filter_negative_numbers(sensor_data)

print("Cleaned sensor data (after filtering negative values):", cleaned_data)
````

### Task 2: String Character Analysis

To build a text-analysis feature, we will create a Python function that counts the number of vowels, consonants, and digits within a given string. We will then demonstrate its usage with sample inputs.

````python
def analyze_string_characters(input_string):
    """
    Counts the number of vowels, consonants, and digits in a string.

    Args:
        input_string (str): The string to analyze.

    Returns:
        dict: A dictionary containing the counts of 'vowels', 'consonants', and 'digits'.
    """
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0

    for char in input_string:
        if char.isalpha(): # Check if the character is an alphabet
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
        elif char.isdigit(): # Check if the character is a digit
            digit_count += 1

    return {
        "vowels": vowel_count,
        "consonants": consonant_count,
        "digits": digit_count
    }
````

````python
# Sample inputs
sample_string_1 = "Hello World 123!"
sample_string_2 = "Python is fun 4 All"
sample_string_3 = "AI & ML"

# Analyze and print results for sample_string_1
result_1 = analyze_string_characters(sample_string_1)
print(f"Analysis of '{sample_string_1}': {result_1}")

# Analyze and print results for sample_string_2
result_2 = analyze_string_characters(sample_string_2)
print(f"Analysis of '{sample_string_2}': {result_2}")

# Analyze and print results for sample_string_3
result_3 = analyze_string_characters(sample_string_3)
print(f"Analysis of '{sample_string_3}': {result_3}")
````

## Generate Palindrome Function (Gemini)

### Subtask:
Generate a Python function for checking palindromes, attributed to 'Gemini', and include example usage with test cases.


**Reasoning**:
The subtask requires defining a Python function to check for palindromes, attributed to 'Gemini', and then demonstrating its usage with test cases. This code block will implement the function and the test cases.



````python
def is_palindrome_gemini(s):
    """
    Checks if a given string is a palindrome, ignoring case and non-alphanumeric characters.
    Attributed to Gemini.

    Args:
        s (str): The input string.

    Returns:
        bool: True if the string is a palindrome, False otherwise.
    """
    # Convert to lowercase and remove non-alphanumeric characters
    cleaned_s = ''.join(char for char in s if char.isalnum()).lower()

    # Compare the cleaned string with its reverse
    return cleaned_s == cleaned_s[::-1]

# Test cases
test_cases = [
    "madam",
    "racecar",
    "A man, a plan, a canal: Panama",
    "hello",
    "Python",
    "Was it a car or a cat I saw?",
    "No lemon, no melon",
    "12321",
    "openai",
    "",
    "a"
]

print("Palindrome Checker (Attributed to Gemini):")
print("-" * 40)
for text in test_cases:
    result = is_palindrome_gemini(text)
    print(f"'{text}': {'Is a palindrome' if result else 'Is NOT a palindrome'}")
````
